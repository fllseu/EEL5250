```python
# Demonstrate instantaneous power
import math
import matplotlib.pyplot as plt
```

```python
# define time as a list of 1000 sample
dt = 0.0001 # 0.0167 is a cycle, so 0.008, 
t_end = 0.2; 
nn = int(t_end/dt)
nn
```

```python
N = range(nn)
tt = [i*dt for i in N]
```

```python
tt
```

```python
N
```

```python
volt = [5*math.cos(377*t) for t in tt]
```

```python
plt.plot(tt, volt)
plt.xlabel('Time (s)')
plt.ylabel('Volt')
plt.show()
```

```python
curr = [1*math.cos(377*t-math.pi/180*60) for t in tt]
p=[]
for i in N:
    p.append(volt[i]*curr[i])

    
```

```python
plt.plot(tt,p)
plt.plot(tt,volt)
plt.plot(tt,curr)
```

```python
plt.show()
```

```python
# root mean square
n = int(0.1/dt); 
rms =[];
for i in range(0, nn-n):
    ms = 0
    for j in range(n):
        ms = ms + volt[i+j]*volt[i+j]   
    ms = ms/n
    rms.append(math.sqrt(ms))
```
